```
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
```

```
t_c = [0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```
def model(t_u,w2,w1,b):
  return w2 * t_u ** 2 + w1 * t_u + b
def loss_funct(t_p, t_c):
    mse = torch.mean((t_p - t_c) ** 2)
    rmse = torch.sqrt(mse)
    return rmse
```

```
parameters = torch.tensor([1.0, 0.0, 0.0], requires_grad=True)
firstlearningrate = 0.0001
optimizer_sgd = optim.SGD([parameters], lr=firstlearningrate)
```

```
def training_loop(n_epochs, optimizer, t_u, t_c, parameters):
    for epoch in range(n_epochs):
        optimizer.zero_grad()
        t_p = model(t_u, *parameters)
        loss = loss_funct(t_p, t_c)
        loss.backward()
        optimizer.step()

        if (epoch + 1) % 500 == 0:
            print(f"Epoch {epoch + 1}, Loss {float(loss)}")
    return parameters
```

```
n_epochs = 5000
training_loop(n_epochs, optimizer_sgd, t_u, t_c, parameters)
print("SGD Optimizer Result:", parameters)
```

```
parameter1 = torch.tensor([1.0, 0.0, 0.0], requires_grad=True)
optimizer_adam = optim.Adam([parameter1], lr=firstlearningrate)
```

```
n_epochs = 5000
t_parameter = training_loop(n_epochs, optimizer_adam, t_u, t_c, parameter1)
print("ADAM Optimizer Result:", t_parameter)
```

```
x_tensor = torch.tensor(x,dtype=torch.float64)
y_tensor = torch.tensor(y,dtype=torch.float64)
```

```
n_samples = x_tensor.shape[0]
n_val = int(0.2*n_samples)
shuffled_indices = torch.randperm(n_samples)
train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]
train_x = x_tensor[train_indices]
train_y = y_tensor[train_indices]
val_x = x_tensor[val_indices]
val_y = y_tensor[val_indices]
```

```
def new_model(x,b,w1,w2,w3,w4,w5):
  predicted = x[:, 0] * b + x[:, 1] * w1 + x[:, 2] * w2 + x[:, 3] * w3 + x[:, 4] * w4 + x[:,5]*w5
  return predicted
def new_loss(pred,actual):
  mse = torch.mean((pred - actual) ** 2)
  rmse = torch.sqrt(mse)
  return rmse
```

```
def new_training_loop(n_epochs, optimizer, params, train_x, train_y, val_x, val_y):
    train_loss_list = []
    val_loss_list = []

    for epoch in range(1, n_epochs + 1):
        print(f"Starting Epoch {epoch}/{n_epochs}...")
        train_pred = new_model(train_x, *params)
        train_loss = new_loss(train_pred, train_y)
        val_pred = new_model(val_x, *params)
        val_loss = new_loss(val_pred, val_y)
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        # Record the training and validation losses
        train_loss_list.append(train_loss.item())
        val_loss_list.append(val_loss.item())
        if epoch % 500 == 0:
            print(f"Epoch {epoch}/{n_epochs} - Training Loss: {train_loss.item():.4f}, Validation Loss: {val_loss.item():.4f}")

    return params, train_loss_list, val_loss_list
```
